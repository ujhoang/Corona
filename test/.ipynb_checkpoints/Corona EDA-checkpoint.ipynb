```python
import pandas as pd
import numpy as np
import cufflinks as cf
cf.set_config_file(offline=True, theme='solar')
```

```python
def read_and_clean_data(name):
    link = "./data/raw/time_series_covid19_{}_global_narrow.csv".format(name)
    df = pd.read_csv(link, skiprows=[1], low_memory=False)
    df["Date"] = pd.to_datetime(df["Date"])
    df_ = df.sort_values("Date", ascending=True)
    df_["Province/State"] = df_["Country"].
    df_final = df_.drop("Province/State", axis = 1)
    df_final.columns = [
        "country",
        "lat",
        "long",
        "date",
        "value_{}".format(name),
        "iso_code",
        "region_code",
        "subregion_code",
        "int_region-code"
    ]
    return df_final
```

```python
df_con = read_and_clean_data("confirmed")
df_dth = read_and_clean_data("deaths")
df_rec = read_and_clean_data("recovered")
```

```python
def merge_dfs(list_of_dfs):
    df_main = list_of_dfs[0].copy()
    for df in list_of_dfs[1:]:
        df_main = pd.merge(df_main, df, on=[
            "country",
            "lat",
            "long",
            "date",
            "iso_code",
            "region_code",
            "subregion_code",
            "int_region-code",
        ])
    return df_main
```

```python
df_main = merge_dfs([df_con, df_dth, df_rec])
```
